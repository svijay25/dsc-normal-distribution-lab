## A quick refresher! 

Here's the formula for the normal distribution density function once more:

$$ \large N(x) = \dfrac{1}{\sigma \sqrt {2\pi }}e^{-\dfrac{(x-\mu)^2}{2\sigma^2}}$$

Here, 
- $\mu$ is the mean
- $\sigma$ is the standard deviation
- $\pi \approx 3.14159 $ 
- $ e \approx 2.71828 $


## First generate a normal distribution containing 5000 values with $\mu=14$ and $\sigma = 2.8$

```python
# Generate a random normal variable with given parameters , n=5000
import numpy as np

# Step 1: Define the parameters
mu = 14
omega = 2.8
num_values = 5000

# Step 2: Generate the normal distribution
normal_dist = np.random.normal(loc=mu, scale=omega, size=num_values)

# Display the first 10 values to verify
print(normal_dist[:10])
```

## Calculate a normalized histogram for this distribution in matplotlib, with bin size = 20

Make sure to get the bin positions and counts for each of the obtained bins. You can use [official documentation](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.hist.html) to view input and output options for `plt.hist()`

```python
# Calculate a histogram for above data distribution
import numpy as np
import matplotlib.pyplot as plt

# Step 2: Calculate the histogram with normalization
counts, bin_edges, _ = plt.hist(normal_dist, bins=20, density=True, alpha=0.6, color='g')

# Step 3: Extract bin positions (midpoints) and counts
bin_positions = (bin_edges[:-1] + bin_edges[1:]) / 2

# Print the bin positions and counts
print("Bin Positions (Midpoints):", bin_positions)
print("Counts (Normalized):", counts)

# Plot the histogram
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Normalized Histogram of Normal Distribution')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

## Use the formula to calculate the density function with $\mu$, $\sigma$ and bin information obtained before

```python
# Calculate the normal Density function 
# Step 3: Calculate the density function values using the PDF function
# Step 1: Define the PDF function for the normal distribution
def normal_pdf(x, mu, omega):
    return (1 / (omega * np.sqrt(2 * np.pi))) * np.exp(-((x - mu) ** 2) / (2 * omega ** 2))

# Step 2: Generate the normal distribution (if not already done)
mu = 14
omega = 2.8
num_values = 5000
normal_dist = np.random.normal(loc=mu, scale=omega, size=num_values)

# Calculate the histogram with normalization
counts, bin_edges, _ = plt.hist(normal_dist, bins=20, density=True, alpha=0.6, color='g')

# Extract bin positions (midpoints)
bin_positions = (bin_edges[:-1] + bin_edges[1:]) / 2

density_values = normal_pdf(bin_positions, mu, omega)

# Print the density values
print("Density Values:", density_values)

# Step 4: Plot the histogram and the density function
plt.figure(figsize=(8, 5))
plt.hist(normal_dist, bins=20, density=True, alpha=0.6, color='g', label='Normalized Histogram')
plt.plot(bin_positions, density_values, color='blue', linestyle='--', label='Density Function')

# Customize the plot
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Normalized Histogram and Density Function of Normal Distribution')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)

# Display the plot
plt.show()
```

## Visualize the distribution using seaborn and plot the KDE

```python
# Use seaborn to plot the histogram with KDE
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Generate the normal distribution (if not already done)
mu = 14
omega = 2.8
num_values = 5000
normal_dist = np.random.normal(loc=mu, scale=omega, size=num_values)

# Step 2: Plot the histogram and KDE using Seaborn
plt.figure(figsize=(10, 6))
sns.histplot(normal_dist, bins=20, kde=True, color='skyblue', stat='density', edgecolor='black')

# Customize the plot
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Histogram and KDE of Normal Distribution')
plt.grid(True, linestyle='--', alpha=0.7)

# Display the plot
plt.show()
```
